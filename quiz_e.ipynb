```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv("Time_series.csv")
data
```

```python
data["Dates"] = pd.to_datetime(data["Tmax"])

data.dtypes
```

```python
print(data.shape)
print(data.head())
data.tail()
```

```python
# replacing ? with NaN
data.replace('?',np.nan, inplace=True)
data.tail(365)
```

```python
data.describe(include='all')
```

```python
# Converting the dtypes of all columns to float32
columns = list(data.columns)


for col in columns:
    data[col] = data[col].astype(str).astype('float32')
data.dtypes
```

```python
#Filling missing values
data.fillna(data.mean(),inplace=True)
data.isnull().sum()
```

```python
# Resampling "Daily"
df_daily = data.resample('D').sum()
```

```python
# Plotting
plt.plot(df_daily.index,df_daily.ElecPrice)
```

```python
# Checking the correlation
data.corr()
```

```python
df_daily.reset_index(inplace=True)

x_test = df_daily.tail(2757)
x_train = df_daily[:-2757]

df_daily.dtypes
```

```python
# Preparing the data for univariate analysis
df_uni = df_daily[['FullDate','ElecPrice']]
df_uni
```

```python
from statsmodels.tsa.stattools import adfuller
adf_result = adfuller(data[ 'SysLoad'])
print( f'ADF Statistic: {adf_result[ 0 ]} ' )
print( f'p-value: {adf_result[ 1 ]} ' )
print( f'No. of lags used: {adf_result[ 2 ]} ' )
print( f'No. of observations used : {adf_result[ 3 ]} ' )
print( 'Critical Values:' )
for k, v in adf_result[ 4 ].items():
    print( f' {k} : {v} ' )
```

```python
new_df = df_uni.rename(columns={"FullDate": "ds", "ElecPrice": "y"})
from fbprophet import Prophet
x_test_uni = new_df.tail(2757)
x_train_uni = new_df[:-2757]
```

```python
# building the model
model = Prophet()
```

```python
# fitting the model
model.fit(x_train_uni)
```

```python
# predicting the test data

future = model.make_future_dataframe(periods=365 , freq= 'D' )
result_uni = model.predict(future)
result_uni.shape
```

```python
result_uni.head()
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
# Plotting the Graph
model.plot(result_uni)
plt.show()
```

```python
# MAPE Value
mean_absolute_percentage_error(x_test_uni['y'], result_uni['yhat'][-2757:])

```

```python
# RMSE Calculation
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(x_test_uni['y'], result_uni['yhat'][-365:]))
print(rmse)

```
